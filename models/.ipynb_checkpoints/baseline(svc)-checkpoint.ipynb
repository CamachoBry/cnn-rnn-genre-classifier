```python
import pandas as pd
import os 
import ast

import sklearn as skl
import sklearn.utils, sklearn.preprocessing, sklearn.decomposition, sklearn.svm
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
import numpy as np
```

Load data

```python
filepath = '../data/fma_metadata/tracks.csv'
tracks = pd.read_csv(filepath, index_col=0, header=[0, 1])
tracks.head()
```

```python
keep_cols =  [('set', 'split'), ('set', 'subset'),('track', 'genre_top'), ('track', 'genres')]
tracks = tracks[keep_cols]
tracks.head()
```

Analyzing MFCC features

```python
filepath = '../data/fma_metadata/features.csv'
features = pd.read_csv(filepath, index_col=0, header=[0, 1, 2], skip_blank_lines=True )
```

```python
mfcc = features['mfcc']
mfcc.head()
```

```python
small = tracks['set', 'subset'] == 'small'

train = tracks['set', 'split'] == 'training'
val = tracks['set', 'split'] == 'validation'
test = tracks['set', 'split'] == 'test'

y_train = tracks.loc[small & train, ('track', 'genre_top')]
y_val = tracks.loc[small & val, ('track', 'genre_top')]
y_test = tracks.loc[small & test, ('track', 'genre_top')]

X_train_mfcc = features.loc[small & train, 'mfcc']
X_val_mfcc = features.loc[small & val, 'mfcc']
X_test_mfcc = features.loc[small & test, 'mfcc']

print(f'Training examples: {y_train.size}')
```

Standarize the features

```python
#Shuffle training features
X_train, y_train = skl.utils.shuffle(X_train_mfcc, y_train, random_state=42)

#Standard Scaling features
scaler = skl.preprocessing.StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val_mfcc)
X_test = scaler.transform(X_test_mfcc)

print(X_train.shape, y_train.shape)
```

```python
#Create a label encoder
le = skl.preprocessing.LabelEncoder()

#Fit transform all y's
y_train = le.fit_transform(y_train)
y_val = le.fit_transform(y_val)
y_test = le.fit_transform(y_test)
```

Building the baseline model

```python
from sklearn.metrics import f1_score
from time import time
```

```python
def train_classifier(clf, X_train, y_train):
    #Start clock, train model, end clock
    start = time()
    clf.fit(X_train, y_train)
    end = time()
    
    print(f'Elapsed training time: {end - start}')
```

```python
def predict_labels(clf, features, target):
    
    start = time()
    y_pred = clf.predict(features)
    end = time()
    
    print(f'Elapsed prediction time: {end - start}')
    return f1_score(target, y_pred, average='micro', pos_label=1)
```

```python
def train_predict(clf, X_train, y_train, X_test, y_test):
    
    #Print classifier being used
    print(f'Training a {clf.__class__.__name__}')
    
    
    #Train the classifier
    train_classifier(clf, X_train, y_train)
    
    #Result of predictions
    print(f'F1 Score for training set: {predict_labels(clf, X_train, y_train)}')
    print(f'F1 Score for testing set: {predict_labels(clf, X_test, y_test)}\n')
          
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

#Initialize instances of all models
clf_dtc = DecisionTreeClassifier(random_state=10, max_depth=4)
clf_svc = SVC()
clf_lr = LogisticRegression(solver='newton-cg')
clf_rfc = RandomForestClassifier(random_state=10, max_depth=30, n_estimators=300, min_samples_leaf=6, min_impurity_decrease=0.0002,
                     class_weight='balanced')
```

```python
for clf in [clf_dtc, clf_svc, clf_lr, clf_rfc]:
    print(f'Model: {clf.__class__.__name__}')
    train_predict(clf, X_train, y_train, X_test, y_test)
    
```

```python
#Analyzing the best classifier (SVC)

from sklearn.metrics import classification_report

y_pred = clf_svc.predict(X_test)
labels = [0, 1, 2, 3, 4, 5, 6, 7]
target_names = le.inverse_transform(labels)

print(classification_report(y_test, y_pred, target_names=target_names))
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))
```
