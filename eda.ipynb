```python
import pandas as pd
import os
import ast

import sklearn as skl
import sklearn.utils, sklearn.preprocessing, sklearn.decomposition, sklearn.svm
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
def load_data(filepath):
    
    filename = os.path.basename(filepath)
    
    if 'features' in filename:
        return pd.read_csv(filepath, index_col=0, header=[0, 1, 2])
    if 'echohonest' in filename:
        return pd.read_csv(filepath, index_col=0, header=[0, 1, 2])
    if 'genres' in filename:
        return pd.read_csv(filepath, index_col=0)
    if 'tracks' in filename:
        tracks = pd.read_csv(filepath, index_col=0, header=[0, 1])
        
        #Changing all string columns into literals using AST
        cols = [('track', 'tags'), ('album', 'tags'), ('artist', 'tags'),
                   ('track', 'genres'), ('track', 'genres_all'),
                   ('track', 'genres_top')]
        
        for column in cols:
            tracks[column] = tracks[column].map(ast.literal_eval)
            
        #Changing all date columns into datetime format
        cols = [('track', 'date_created'), ('track', 'date_recorded'),
                   ('album', 'date_created'), ('album', 'date_released'),
                   ('artist', 'date_created'), ('artist', 'active_year_begin'),
                   ('artist', 'active_year_end')]
        
        for column in cols:
            tracks[column] = pd.to_datetime(tracks[column])
            
        #Converting to categorical data    
        subsets = ('small', 'medium', 'large')
        tracks['set', 'subset'] = tracks['set', 'subset'].astype('category', 
                                                                 categories=subsets, ordered=True)
        
        cols = [('track', 'license'), ('artist', 'bio'),
                   ('album', 'type'), ('album', 'information')]
        for column in cols:
            tracks[column] = tracks[column].astype('category')
        
        return tracks
```

Exploring Features CSV

```python
filepath = 'data/fma_metadata/features.csv'
features = pd.read_csv(filepath, index_col=0, header=[0, 1, 2], skip_blank_lines=True)
features.head()
```

```python
mfcc = features['mfcc']
mfcc.head()
```

```python
#Used for boolean masking
small = tracks['set', 'subset'] <= 'small'
genre1 = tracks['track', 'genre_top'] == 'Electronic'
genre2 = tracks['track', 'genre_top'] == 'Instrumental'


X = features.loc[small & (genre1 | genre2), 'mfcc']
X = skl.decomposition.PCA(n_components=2).fit_transform(X)

y = tracks.loc[small & (genre1 | genre2), ('track', 'genre_top')]
y = skl.preprocessing.LabelEncoder().fit_transform(y)

#Graph mfcc features to see if there is any visualized distinction
plt.scatter(X[:,0], X[:,1], c=y, cmap='RdBu', alpha=0.5)
plt.show()
```
